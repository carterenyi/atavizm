```python
import pandas as pd
import mido
```

```python
mid = mido.MidiFile("../Data/Fugue-BWV-847.mid")
```

```python
mid
```

```python
mid.length
```

```python
for i, track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i, track.name))
    for msg in track:
        if not msg.is_meta:
            print(msg)
```

```python
midiframe = pd.DataFrame(columns=["Voice", "Pitch", "Velocity", "Ticks"])
```

```python
for i, track in enumerate(mid.tracks):
    for msg in track:
        if msg.type == 'note_on':
            midiframe.Voice = i
            midiframe.Pitch = msg.note
            midiframe.Velocity = msg.velocity
            midiframe.Ticks = msg.time
midiframe.head()
```
